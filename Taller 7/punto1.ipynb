```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

X = np.array([[2, -1],[1, 2], [1, 1]])
y = np.array([[2], [1], [4]])

def puntos (X,y):
  X = np.array(X)
  y = np.array(y)
  XT = np.dot(X.T,X)
  yT = np.dot(X.T,y) 
  xsol = np.linalg.solve(XT,yT)
  return xsol

resultado=puntos(X,y)
print (resultado)
```

```python
x = np.linspace(0,10,100)
y1 = 2*x-2
y2 = (1-x)/2
y3 = 4-x

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1, 1, 1)
plt.plot(x,y1,c = 'b')
plt.plot(x,y2,c = 'green' )
plt.plot(x,y3, c= 'r')
plt.scatter(resultado[0],resultado[1],c='c')
plt.show()

print('El método de mínimos cuadrados calculó el punto mas cercano a las 3 rectas, ya que no existe un punto por el cual pasen las 3')
```
